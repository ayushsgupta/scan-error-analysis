```python
A = np.load('data/A.npy')
A0 = np.load('data/A0.npy')
b_abs = np.load('data/b_abs.npy')
b_rel = np.load('data/b_rel.npy')
```

```python
num_test = 20
num_train = len(b_abs) - num_test
```

```python
A_train, A_test = A[:num_train], A[num_train:]
A0_train, A0_test = A0[:num_train], A0[num_train:]
ba_train, ba_test = b_abs[:num_train], b_abs[num_train:]
br_train, br_test = b_rel[:num_train], b_rel[num_train:]
```

```python
print(A_train.shape, ba_train.shape, br_train.shape)
```

```python
models = ['LinearRegression', 
          'ElasticNet', 
          'DecisionTreeRegressor', 
          'RandomForestRegressor',
          'GradientBoostingRegressor']
model_classes = [eval(m) for m in models]
```

```python
def model_prediction(model_class, A_train, b_train, A_test):
    model = model_class()
    model.fit(A_train, b_train)
    return model.predict(A_test)

def model_evaluation(model, metric, rel=True, filled=True):
    Atr = A_train if filled else A0_train
    Ate = A_test if filled else A0_test
    b = br_train if rel else ba_train
    btest = br_test if rel else ba_test
    prediction = model_prediction(model, Atr, b, Ate)
    return metric(btest, prediction)
```

```python
def annotated_matrix_heatmap(data, xlabels, ylabels, metricname): 
    # https://matplotlib.org/3.1.1/gallery/images_contours_and_fields/image_annotated_heatmap.html
    farmers = xlabels
    vegetables = ylabels
    harvest = data
    
    plt.rcParams['figure.figsize'] = [12, 12]

    fig, ax = plt.subplots()
    im = ax.imshow(harvest)

    # We want to show all ticks...
    ax.set_xticks(np.arange(len(farmers)))
    ax.set_yticks(np.arange(len(vegetables)))
    # ... and label them with the respective list entries
    ax.set_xticklabels(farmers)
    ax.set_yticklabels(vegetables)

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    for i in range(len(vegetables)):
        for j in range(len(farmers)):
            text = ax.text(j, i, harvest[i, j], ha="center", va="center", color="w")

    ax.set_title("Logarithmic error quantified by {} for various models and parameters".format(metricname))
#     fig.tight_layout()
    plt.show()
```

```python
def compare_on_metric(metric, log=True):
    harvest = []
    ylabels = models
    xlabels = ['Rel=T,Fill=T','Rel=T,Fill=F','Rel=F,Fill=T','Rel=F,Fill=F']
    for m in model_classes:
        row = []
        for param in [{'rel':True, 'filled':True}, 
                      {'rel':True, 'filled':False}, 
                      {'rel':False, 'filled':True}, 
                      {'rel':False, 'filled':False}]:
            if log:
                row.append(np.log(model_evaluation(m, metric, param['rel'], param['filled'])))
            else:
                row.append(model_evaluation(m, metric, param['rel'], param['filled']))
        harvest.append(row)
    annotated_matrix_heatmap(np.array(harvest), xlabels, ylabels, metric.__name__)
```

```python
compare_on_metric(mean_squared_error)
```

```python
compare_on_metric(max_error)
```

```python
compare_on_metric(mean_absolute_error)
```
