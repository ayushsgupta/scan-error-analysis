```python
b = np.load('data/b.npy')
b = np.abs(b) > 0.04
```

```python
len(b) == len(list(bader_statistics.keys()))
```

```python
idx = 0
cation_low_error, cation_high_error, anion_low_error, anion_high_error = [], [], [], []

for ID in bader_statistics.keys():
    ncats, nans = 0, 0
    formula = binary_ids_formulas[int(ID)]
    cation, anion = [str(e) for e in Composition(formula).elements]
    avg_stats = dict()
    
    for well in bader_statistics[ID]:
        ion = list(well.keys())[0]
        if ion == cation:
            avg_stats[ion] = avg_stats.get(ion, 0) + well[ion][0]
            ncats += 1
        elif ion == anion:
            avg_stats[ion] = avg_stats.get(ion, 0) + well[ion][0]
            nans += 1
    avg_stats[cation] = avg_stats[cation] / ncats
    avg_stats[anion] = avg_stats[anion] / nans
    
    diffc = avg_stats[cation] - element_statistics[cation][0]
    diffa = avg_stats[anion] - element_statistics[anion][0]
    
    if b[idx]:
        cation_high_error.append(diffc)
        anion_high_error.append(diffa)
    else:
        cation_low_error.append(diffc)
        anion_low_error.append(diffa)

    idx += 1
```

```python
logify = lambda lst: [np.log(elem) if elem > 0 else -np.log(-elem) for elem in lst]
```

```python
#Adapted from pymatgen's pretty_plot, but with some custom specs
def plot_helper(plot, 
                width=8, 
                height=None, 
                dpi=None, 
                color_cycle=("qualitative", "Set1_9"),
                title=None,
                xlabel=r'$log(\Delta \alpha)$',
                ylabel='Count'):
    
    ticksize = int(width * 2)
    golden_ratio = (np.sqrt(5) - 1) / 2
    if not height:
        height = int(width * golden_ratio)

    fig = plot.gcf()
    fig.set_size_inches(width, height)
    plot.xticks(fontsize=ticksize)
    plot.yticks(fontsize=ticksize)

    ax = plot.gca()

    plot.title(title)
    ax.set_title(ax.get_title(), size=width * 2.5, pad=25)

    labelsize = int(width * 1.7)
    plot.xlabel(xlabel, size=ticksize*1.1)
    plot.ylabel(ylabel, size=ticksize*1.1)

#     ax.set_xlabel(ax.get_xlabel(), size=labelsize)
#     ax.set_ylabel(ax.get_ylabel(), size=labelsize)
    handles, labels = ax.get_legend_handles_labels()
    hl = sorted(zip(handles, labels), key=lambda x: x[1])
    handles2, labels2 = zip(*hl)
    plt.legend(handles2, labels2, prop={'size': 1.7 * width}, edgecolor='black')
    
    plt.axvline(x=0.00, color='black', lw=2, linestyle='--')

    return plot
```

```python
binz, width = 40, 32

plt.hist([logify(cation_high_error), logify(anion_high_error)], 
         bins=binz, 
         histtype='bar', 
         label=['Cations', 'Anions'],
         color=['#006AFF', '#39CC6E']
)
plot = plot_helper(plt, 
            width=15,
            title=r'High Error Compounds: $\alpha_{Bonded} - \alpha_{Element}$')
plot.show()
```

```python
binz, width = 40, 32

plt.hist([logify(cation_low_error), logify(anion_low_error)], 
         bins=binz, 
         histtype='bar', 
         label=['Cations', 'Anions'],
         color=['#006AFF', '#39CC6E']
)
plot = plot_helper(plt, 
            width=15,
            title=r'Low Error Compounds: $\alpha_{Bonded} - \alpha_{Element}$')
plot.show()
```
