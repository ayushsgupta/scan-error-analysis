```python
hfkey = list(thermodata[0]['data']['fcc']['298K'].keys())[0]
# print(hfkey)

duplicates = set()

for p in thermodata:
    phases = [phase for phase in p['data'].keys()]
    energies = [p['data'][phase]['298K'].get(hfkey, '50000000') for phase in phases]
    references = [p['data'][phase]['reference'] for phase in phases]
    for e in itertools.combinations(energies, 2):
        p0, p1 = phases[energies.index(e[0])], phases[energies.index(e[1])]
        r0, r1 = references[energies.index(e[0])], references[energies.index(e[1])]
        e0, e1 = float(re.sub(r'[a-zA-Z\/]+', '', e[0])), float(re.sub(r'[a-zA-Z\/]+', '', e[1]))
        if approx(e0, e1) and r0 != r1:
            duplicates.add(p['identifier'] + ' ' + str(thermodata.index(p)) + ' ' + p0 + ' ' + str(e0) + ' ' + p1 + ' ' + str(e1))
```

```python
list(duplicates)
```

```python
with open('duplicates.txt', 'w') as filehandle:
    for item in duplicates:
        filehandle.write('%s\n' % item)
```
